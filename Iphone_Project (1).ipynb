```python
import pandas as pd
```

```python
df1=pd.read_csv('iphone.csv')
```

```python
df1.head()
```

```python
#1. The column names have spaces . rename the column names to have underscore '_' instead of space (try to do in one go instead of specifying each column nam in rename method)
df1.columns=df1.columns.str.replace(' ','_')
```

```python
df1.columns
```

```python
#2- start rating for some of the models is missing in the dataset. fill those missing values with the average rating all the models.
avg_rating=df1['Star_Rating'].mean()

# Fill Na
df1['Star_Rating'].fillna(avg_rating,inplace=True)
```

```python
#3- Now instead of filling missing values with avg rating of full dataset ,
#fill with avg rating based on RAM. example :  if rating for a 2 gb phone is
#missing then take average of all other 2 gb phones rating and fill that value.

average_ratings=df1.groupby('Ram')['Star_Rating'].mean().reset_index()
```

```python
average_ratings.rename(columns = {'Star_Rating' : 'Average Rating'}, inplace=True)
df1 = df1.merge(average_ratings, on='Ram', how='left')
```

```python
df1['Star_Rating'].fillna(df1['Average Rating'],inplace=True)
df1.drop(columns=['Average Rating'])
```

```python
#4- create a new column in the dataframe "Discount_Percentage" based on MRP and sale value
df1['Discount_Percentage']=((df1['Mrp'] - df1['Sale_Price']) / df1['Mrp']) * 100
```

```python
#5- which model has highest percent discount ?
sort_discount_percentage=df1.sort_values(by='Discount_Percentage',ascending=False)
model_highest_discount=sort_discount_percentage.iloc[0]['Product_Name']
highest_percent=sort_discount_percentage.iloc[0]['Discount_Percentage']
print(model_highest_discount)
print(highest_percent)
```

```python
#6- find total no of models  each space configuration (128 GB , 64 GB etc)
df1.groupby('Ram')['Product_Name'].count()
```

```python
#7- list top 5 models having highest no of reviews

# Sort the DataFrame by 'Number Of Reviews' in descending order
df_sorted = df1.sort_values(by='Number_Of_Reviews',ascending=False	)

# Select the top 5 models with the highest number of reviews
df_sorted.head(5)
```

```python
#8 - what is the price diffrence between highest price and lowest price iphone (based on mrp)
highest_mrp=df1['Mrp'].max()
lowest_mrp=df1['Mrp'].min()
price_diff=highest_mrp-lowest_mrp
print(price_diff)
```

```python
#9- find total no of reviews for iphone 11 and iphone 12 category . Output should have only 2 rows (for 11 and 12).
df1.loc[df1['Product_Name'].str.contains('iPhone 11',case=False),'iphone Category']='iPhone 11'
df1.loc[df1['Product_Name'].str.contains('iPhone 12',case=False),'iphone Category']='iPhone 12'
df1.groupby('iphone Category')['Number_Of_Reviews'].count()
```

```python
#10 which iphone has 3rd highest MRP
mrp_sort=df1.sort_values(by='Mrp',ascending=False)
#mrp_sort
mrp_sort.iloc[2]['Product Name']
```
